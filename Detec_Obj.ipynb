```python
# ver a img
import cv2 # lib opencv
from google.colab.patches import cv2_imshow
# add variavem detector_carros
detector_carros = cv2.CascadeClassifier('/content/drive/MyDrive/Visão Computacional Guia Completo/Cascades/cars.xml')
imagem = cv2.imread('/content/drive/MyDrive/Visão Computacional Guia Completo/Images/car.jpg')

if imagem is None:
  print('Img not found')
else:
  imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
  carros = detector_carros.detectMultiScale(imagem_cinza, scaleFactor=1.01, minNeighbors=5, minSize=(20,20) ) # pega o tamanho minimo dos obj

  print(f"Tot the car detect: {len(carros)}")
  print("Coordenadas the car: ", carros)

for (x, y, w, h) in carros:
  cv2.rectangle(imagem, (x, y), (x + w, y + h), (0, 0, 255), 2)
  #ver tamanhos dos carros
  len(carros)
  print(carros)
cv2_imshow(imagem)
```
