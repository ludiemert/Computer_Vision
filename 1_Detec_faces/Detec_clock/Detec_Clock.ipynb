```python
import cv2
from google.colab.patches import cv2_imshow
detec_clock = cv2.CascadeClassifier('/content/drive/MyDrive/Visão Computacional Guia Completo/Cascades/clocks.xml')
img = cv2.imread('/content/drive/MyDrive/Visão Computacional Guia Completo/Images/clock.jpg')

if img is None:
    print('Img not found')
else:
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    clock = detec_clock.detectMultiScale(gray, scaleFactor = 1.03, minNeighbors=1,) # minSize=(10, 10)

    print(f"Tot Clock: {len(clock)}")
    print("Coordenadas the clock: ", clock)

    for (x, y, w, h) in clock:
      cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
      #roi_gray = gray[y:y + h, x:x + w]
      #roi_color = img[y:y + h, x:x + w]
    len(clock)
    print(f"Tot Clock: {len(clock)}")
    cv2_imshow(img)
```
