```python
import cv2
from google.colab.patches import cv2_imshow
full_body = cv2.CascadeClassifier('/content/drive/MyDrive/Visão Computacional Guia Completo/Cascades/fullbody.xml')
img = cv2.imread('/content/drive/MyDrive/Visão Computacional Guia Completo/Images/people3.jpg')

if img is None:
    print('Erro ao carregar a imagem.')
else:
  gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  bodies = full_body.detectMultiScale(gray, scaleFactor= 1.05, minNeighbors=5, minSize=(50, 50))

  print('Num person find:', len(bodies))
  print(bodies)

  for (x, y, w, h) in bodies:
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 255), 2)
    cv2_imshow(img)
```
